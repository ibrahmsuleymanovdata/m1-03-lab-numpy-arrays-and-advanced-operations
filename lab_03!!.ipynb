```python
import numpy as np
rng = np.random.default_rng(seed=13)
sim = rng.normal (0.0,1,size = (1000,6))
```

```python
sim.shape
```

```python
sim.dtype
```

```python
sim[:3]
```

# Task 7

```python
scenario_shift= np.array([-0.3, -0.1, 0.0, 0.1, 0.2, 0.4])
scenario_shift.shape
```

```python
adjusted_sim=sim + scenario_shift # sim's shape is (1000,6); scenario_shift's shape is (6,)
```

```python
print("are their shapes same:", adjusted_sim.shape==sim.shape)
```

```python
print("original column means:", sim.mean(axis=0))
print("adjusted_sim means:", adjusted_sim.mean(axis=0))
print("mean difference:", adjusted_sim.mean(axis=0) - sim.mean(axis=0))
```

# Task 8

```python
scenario_means = adjusted_sim.mean(axis=0)
scenario_means
```

```python
scenario_ranking = np.argsort(scenario_means)
scenario_ranking
```

```python
partitioned_vals = np.partition(scenario_means, -2)
top_two_partition = partitioned_vals[-2:]
partitioned_vals
print("top2:", top_two_partition)
```

```python
full_sorted_vals = np.sort(scenario_means)
top_two_sorted = full_sorted_vals[-2:]
print("Top2(from full sort):", top_two_sorted)
```

```python
print("are They absolutely same?:",top_two_partition == top_two_sorted)
```
